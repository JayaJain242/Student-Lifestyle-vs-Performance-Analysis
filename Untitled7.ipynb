```python
df["category"].value_counts().plot(kind="bar")
```

```python
df.groupby("category")["study hours"].mean().plot(kind="bar")
```

```python
df.groupby("category")["social media hours"].mean().plot(kind="bar")
```

```python
df["category"].value_counts().plot(kind="pie", autopct='%1.1f%%')
```

```python
df.plot(x="study hours", y="marks", kind="scatter")
import matplotlib.pyplot as plt
plt.show()
```

```python
df.plot(x="attendance", y="marks", kind="scatter")
plt.show()
```

## Insights

1. Topper students have the highest average study hours compared to average and low-performing students.
2. Students with higher social media usage tend to have lower academic performance.
3. Average students maintain a balance between study, sleep, and social media usage.
4. There is a positive relationship between study hours and marks, as shown in the scatter plot.
5. Students with higher attendance tend to achieve better marks, highlighting the importance of regular class participation.
6. An exception case was observed where a student with low study hours still achieved high marks, indicating that performance is influenced by multiple factors.
7. The scatter plot shows a general trend but with scattered points, suggesting that performance is not solely dependent on study hours.
